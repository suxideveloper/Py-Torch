```python
x = np.arange(1, 11)
y = x**4 + 9

plt.plot(x, y)
```

```python
y, x
```

```python
plt.plot(x, y, c = 'r', marker = '*', linestyle = '--', markersize = 10, markerfacecolor = 'yellow', markeredgewidth = 2, markeredgecolor = 'blue')
```

```python
plt.plot(x, y, c = 'r', marker = '*', linestyle = '--', markersize = 10, markerfacecolor = 'yellow', markeredgewidth = 2, markeredgecolor = 'blue')
plt.title("y = x^4 + 9")
plt.xlabel("X axis")
plt.ylabel("Y axis")
plt.grid(True)
plt.show()
```
